# [Tesla Stocks Prices dataset](https://www.kaggle.com/datasets/timmofeyy/-tesla-daily-stocks-prices)

* Time series analysis
* Comparison between prices
* Relation Adj Close - Volume

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
```

```python
df = pd.read_csv("datasets/tesla_stocks.csv")
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
df.describe()
```

```python
df.set_index('Date', inplace=True)
df.head()
```

```python
df.index
```

```python
df.index = df.index.astype('datetime64[ns]')
```

```python
df['Adj Close'].plot()
```

```python
df[['Open', 'Close']].resample('Y').mean().plot(kind='bar')
```

```python
df[['High', 'Low']].resample('Y').mean().plot(kind='bar')
```

```python
df['Range'] = df['High'] - df['Low']
df['Range'].plot()
```

```python
df['Month'] = df.index.month
df['Year'] = df.index.year
df.head()
```

```python
plt.rcParams['figure.figsize'] =  (8, 8)
df.drop(['Volume', 'Range', 'Year'], axis=1).groupby('Month').mean().plot(kind='bar')
```

```python
last4years = df['2019':]
last4years.head()
```

```python
last4years[['Open', 'Close']].resample('Y').mean().plot(kind='bar')
```

```python
sns.boxplot(data=last4years, x='Year', y='Adj Close')
```

```python
df['Volume'].plot()
```

```python
plt.scatter(df['Adj Close'], df['Volume'], alpha=0.7)
plt.xlabel('Adj Close')
plt.ylabel('Volume')
```
